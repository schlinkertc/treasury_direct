```python
import requests
import json
import pandas as pd
import numpy as np
```

```python
def response(request):
    response = request.json()
    
    return pd.DataFrame(response)
```

```python
def securities_search_byType(security_type):
    url = 'http://www.treasurydirect.gov/TA_WS/securities/search'
    params = {'type':'{}'.format(security_type)}
    
    r = requests.get(url,params)
    return response(r)
```

```python
bonds = securities_search_byType('Bond')
```

```python
def securities_search_byCUSIP(cusip):
    url = 'http://www.treasurydirect.gov/TA_WS/securities/search?cusip={}&format=json'.format(cusip)
    
    r = requests.get(url)
    return response(r)
```

```python
def securities_searchby_issueDate(startDate,endDate):
    url = 'http://www.treasurydirect.gov/TA_WS/securities/search?startDate={}&endDate={}&format=json&dateFieldName=issueDate'.format(startDate,endDate)
    
    r = requests.get(url)
    return response(r)
```
